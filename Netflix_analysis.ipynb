```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import zipfile

with zipfile.ZipFile('netflix_titles.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('.')
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
print(df.shape)
```

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Movies vs TV Shows')
plt.ylabel('')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='bar', color='red')
plt.title('Content Added Per Year on Netflix')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
df['country'].value_counts().head(10)
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='blue')
plt.title('Top 10 Countries on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

```python
genres = df['listed_in'].str.split(', ').explode()
genres.value_counts().head(10)
```

```python
genres.value_counts().head(10).plot(kind='bar', color='green')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
